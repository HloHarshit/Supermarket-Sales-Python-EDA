## Question 13: Display daily sales by day of the week.

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
dw_mapping = {
    0: 'Mon',
    1: 'Tue',
    2: 'Wed',
    3: 'Thu',
    4: 'Fri',
    5: 'Sat',
    6: 'Sun'
}
```

```python
df['day_of_week'] = df['Date'].dt.dayofweek.map(dw_mapping)
```

```python
df['day_of_week'].value_counts()
```

```python
sns.countplot(x=df['day_of_week'])
plt.show()
```

## Question 14: What will be the highest months for sales?

```python
dm_mapping = {
    1: 'Jan',
    2: 'Feb',
    3: 'Mar',
    4: 'Apr',
    5: 'May',
    6: 'Jun',
    7: 'Jul',
    8: 'Aug',
    9: 'Sep',
    10: 'Oct',
    11: 'Nov',
    12: 'Dec'
}
```

```python
df['Month'] = df['Date'].dt.month.map(dm_mapping)
```

```python
df['Month'].value_counts()
```

```python
sns.countplot(x=df['Month'])
plt.show()
```
